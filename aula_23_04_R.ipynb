### 2. Existe uma relação entre a potencia do motores (hp) e aceleração (qsec) do carro (modelo de regressão linerar)


```r
hp <- df$hp
qsec <- df$qsec
```

```r
modelo <- lm(qsec ~ hp)
summary(modelo)
```

### 3 . quais são os carros mais econômico em termos de consumo de combustivel(mpg)

```r
carros <- data.frame(mpg =df$mpg, modelo = rownames(mtcars))
carros_ordenados <- carros[order(carros$mpg),]

barplot(carros_ordenados$mpg,names.arg = carros_ordenados$modelo, horiz = TRUE,col = "darkblue" ,xlab= "mpg",main = "Consumo de combustivel dos carros")
```
